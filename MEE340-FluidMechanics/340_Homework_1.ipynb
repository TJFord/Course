

## Solution : 
The velocity profile for laminar flow is given. The relation for the drag force applied by the fluid on a section of the pipe and the value of the drag force for the water conditions given must be determined.

## Assumptions:
1. The fluid flow thorugh a circular pipe is one dimentional
2. The fluid is Newtonian
2. The velocity profile equation is $u=u_{max}\cdot(1-\dfrac{r^2}{R^2})$
3. Water flow at $20C$ and with a viscosity of $0.0010 \dfrac{kg}{m \cdot{s}}$

## Properties:
"Viscosity also causes a drag force on a solid surface in contact with the fluid: the viscous stress at the wall, $T_w = (-\mu)\cdot(\dfrac{du}{dr})$, transmits the fluid drag to the surface"

"The drag force on a solid surface can be found from the velocity profile by evaluating the velocity gradient at the wall and integrating the surface stress over the area of the body."

```python
%%render param
m = 3*u.kg #mass
R = 0.08*u.m #radius of the pipe
L = 30*u.m #length of the pipe
u_max = 3*(u.m/u.s) #maximum flow velocity
mu = 0.001*(u.kg/(u.m*u.s)) 
pi = math.pi
```

## Analysis:

1. The velocity profile is given by the equation
\begin{equation*}
u(r) = u_{max}(1-\dfrac{r^2}{R^2})
\end{equation*}
where $R$ is the radius of the pipe, $r$ is the radial distance from the center of the pipe, and $u_{max}$ is the maximum flow velocity, which happens at the center where $r=0$.


2. The shear stress at the pipe surface is given by the expression
\begin{equation*}
\tau_w = -\mu\dfrac{du}{dr}\bigg\rvert_{r=R} = -\mu u_{max}\dfrac{d}{dr}(1-\dfrac{r^2}{R^2})_{r=R} =  -\mu u_{max}\dfrac{-2r}{R^2}\bigg\rvert_{r=R} = \dfrac{2 \mu u_{max}}{R}
\end{equation*}
Because the quantity $\dfrac{du}{dr}$ is negative in pipe flow, the negative sign is added to the $\tau_w$ relation which makes the shear stress in the positive direction.


3. The friction drag force exerted by the fluid on the inner surface of the pipe becomes
\begin{equation*}
F_D = \tau_w A_s = \dfrac{2 \mu u_{max}}{R}(2\pi RL) = 4\pi \mu L u_{max}
\end{equation*}

```python
%%render 
F_D = 4*pi*mu*u_max*L
```

```python
print(F_D.to('N'))
```

The formula to calculate the drag force is obtained from the derivative of the velocity profile, viscuous stress equation, and area to find that $F_D = 1.131 N$ in the condition for water flow at $20C$.
