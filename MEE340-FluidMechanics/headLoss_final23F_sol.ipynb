

$$\dfrac{P_p}{\rho g}+\alpha \dfrac{ V^2}{2g}+z_0= \dfrac{P_t}{\rho g}+d + K_L\dfrac{V^2}{2g}+f\dfrac{L}{D}\dfrac{V^2}{2g} $$
$$\dfrac{P_p}{\rho g}= \dfrac{P_t}{\rho g}+d + K_L\dfrac{V^2}{2g}+f\dfrac{L}{D}\dfrac{V^2}{2g} - \alpha \dfrac{ V^2}{2g} $$

```python
%%render
rho = 1000*u.kg/u.m**3
g = 9.81*u.m/u.s**2
L = 100*u.m
D = 100*u.mm
d = 10*u.m
Q = 0.01*u.m**3/u.s
mu = 1e-3*u.Pa*u.s
a = 1.1
K_L= 1.0
V_p = 4*Q/(np.pi*D**2) # velocity in the pipe
V_t = 0*u.m/u.s # surface velocity at the top of the tank
# Energy equation
P_t = 0*u.Pa
Re = rho*D*V_p/mu
f = 1/(-1.8*np.log(6.9/Re))**2
h_p = d  +(-a+f*L/D+K_L)*V_p**2/2/g
```
