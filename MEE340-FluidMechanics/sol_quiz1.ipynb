# Problem
A 6-m-high, 2-m-wide (into the paper) rectangular plate blocks the end of a 5-m-deep freshwater channel, as shown in the figure. The plate is hinged about a horizontal axis along its upper edge through a point A and is restrained from opening by a fixed ridge at point B. Determine the force exerted on the plate by the ridge. Your solution should contain units. Hint: static pressure $p=\rho g h$. 

# Plan
## Thought process
The problem asks for force $F_B$ at point B. How does that relate to the fluid pressure? Pressure will generate a pressure force $F_p$ on the block, it looks like a free body diagram for the block will help as it connects both pressure force and $F_B$. The block is in equilibrium, so equations of equilibirum can be used to find $F_B$. Don't forget the reaction forces from point A. 


## Solution

The pressure force can be calculated as $$F_p=p_cA$$
where $p_c$ is the pressure at the centroid, $A$ is the area immersed in the fluid. 
$$p_c = \rho g h_c=\rho g\dfrac{\overline{CB}}{2}$$ where the centroid of the immersed rectangle is half of its height, i.e., $h_c=\dfrac{\overline{CB}}{2}$, $A=\overline{CB} W = 5\times 2\, m^2$

The action point $H$ should be $\frac{1}{3}$ of the imemrsed height $5~m$, so $\overline{BH} = 5/3~m$

Apply the moment balance equation at $A$
$$\sum M_A = F_p\overline{AH}-F_B\overline{AB}=0$$
so we can solve it as $$F_B = F_p\dfrac{\overline{AH}}{\overline{AB}}$$

## discussion
The final expression relates $F_B$ to $F_p$, the bigger pressure force, the bigger $F_B$ is required to hold the block. That makes sense. $F_B$ is also proportial to the ratio of $\dfrac{\overline{AH}}{\overline{AB}}$, which is the insight given by analytical expression compared with numerical values. 


```python
from pint import UnitRegistry
u=UnitRegistry(auto_reduce_dimensions=True)
import handcalcs.render
```

```python
%%render short
rho = 1000*u.kg/u.m**3 # water density
g = 9.81*u.m/u.s**2 # gravity constant
CB = 5*u.m # immersed height
W = 2*u.m # width of the block
Area = CB*W # area
h_c = CB/2 # centroid location in height
AC = 1 *u.m # height above water surface measured from A
CH = 2/3*CB # HB is 1/3 CB

AB = 6*u.m # total height
AH = AC + CH

F_p = rho*g*h_c*Area
F_B = F_p*AH/AB
```

```python
print(f'The force F_B is {F_B.to(u.N)}')
print(f'The pressure force F_p is {F_p.to(u.N)}')
```
