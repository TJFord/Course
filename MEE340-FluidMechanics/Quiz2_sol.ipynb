```python
import sympy as sp # use symbolic python taking derivative
import handcalcs.render # render
from sympy.vector import CoordSys3D
import numpy as np # numerical python library
import matplotlib.pyplot as plt
x, y, z, t = sp.symbols('x y z t') # define it as symbols
sp.init_printing(use_unicode=True)
```

```python
%%render
N = CoordSys3D('N') # define cartesian coordinates
V = (x**2-y**2)*N.i-2*x*y*N.j # velocity vector
n1 = N.j # surface norm
n2 = N.i # surface norm
n3 = sp.sqrt(2)/2*N.i - sp.sqrt(2)/2*N.j # surface norm
```

```python
%%render
# problem a
u = V.dot(N.i)
v = V.dot(N.j)
a = sp.diff(V,t) + u*sp.diff(V,x) + v*sp.diff(V,y)
a_23=a.subs({x:2, y: 3})
a_v = a_23.magnitude() # cm/s^2
```

```python
%%render
# problem b and c
# flux from top
exp_t = V.dot(n1) # dot product
exp_ts = exp_t.subs(y,2) # y==2 at the top
Q_t = sp.integrate(exp_ts,(x,0,4),(z,0,1)) # integrate with x
Q_t # cm^3/s

# flux from the right
exp_r = V.dot(n2) # dot product
exp_rs = exp_r.subs(x,4) # x==4 on the right
Q_r = sp.integrate(exp_rs,(y,-2,2),(z,0,1)) # integrare with y
Q_r # cm^3/s


#flux from the diagonal
exp_d = sp.sqrt(2)/2*(x**2-y**2)+sp.sqrt(2)*x*y
exp_ds = exp_d.subs(y,x-2)
Q_d = sp.integrate(exp_ds,(x,0,4),(z,0,1)) # cm^3/s
```

```python
print(f'the volume flow rate from the top surface is {Q_t} cm^3/s')
print(f'the volume flow rate from the right surface is {Q_r} cm^3/s')
print(f'the volume flow rate from the diagonal line is {Q_d} cm^3/s')
```
