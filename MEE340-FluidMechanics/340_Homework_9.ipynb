## Assumptions:
Horizontal force on a fireman's nozzle as a funciton of $V_1$, pressure drop $\bigtriangleup P = P_1 - P_2$, density $\rho$, viscosity $\mu$, inlet area $A_1$, outlet area $A_2$, and length $L$. $F = f(V_1, \bigtriangleup P, \rho, \mu, A_1, A_2, L)$, use $V_1$, $A_1$, and $\rho$ as repeating parameters.

## Analysis:
List parameters: $n = 8$
List primary dimensions of each parameter:
\begin{align*}
\{F\} = \{mass(\dfrac{length}{time^2})\} \\ \{V\} = \{\dfrac{length}{times}\} \\ \{\bigtriangleup P\} = \{\dfrac{mass}{time^2}(length)\} \\ \{\rho\} = \{\dfrac{mass}{length^3}\} \\ \{\mu\} = \{ \dfrac{mass}{length}(time)\} \\ \{A_{1,2}\} = \{length^2\} \\ \{L\} = \{length\}
\end{align*}

Set reduction: j = 3

Set number of dimensionless groups: $K_1 = n - j = 5$

### Analysis for dimensionless group 1:
\begin{equation*}
\Pi_1 = FV_1^{a_1}A_1^{b_1}\rho^{c_1} \\ \{\Pi_1\} = \{(m^1L^1t^-2)(L^1t^-1)^{a_1}(L^2)^{b_1}(m^1L^{-3})^{c_1}\}
\end{equation*}
\begin{equation}
mass: \{m^0\}= \{m^1m^{c_1}\} \qquad 0 = 1 + c_1 \qquad c_1 = -1\\ 
time: \{t^0\}= \{t^{-2}t^{-a_1}\} \qquad 0 = -2 - a_1 \qquad a_1 = -2 \\
length: \{L^0\} = \{L^1L^{a_1}L^{2b_1}L^{-3c_1}\} \qquad 0 = 1 + a_1 + 2b_1 -3c_1 \qquad b = -1 \\
\end{equation}

Plugging in the exponents:
\begin{equation*}
\Pi_1 = \dfrac{F}{\rho V_1^2 A_1}
\end{equation*}

### Analysis for dimensionless group 3:
