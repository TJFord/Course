```python
import handcalcs.render
import sympy as sym
from pint import UnitRegistry
u=UnitRegistry(auto_reduce_dimensions=True)
from sympy import symbols, diff
```


## Solution:
Given the velocity field in the xy-plane, calculate the acceleration field and acceleration at the point $(x ,y) = (-1 ,3)$

## Assumptions:
the velocity field is given with $u$ and $v$ as in $V(u, v)$, where $u = 1.85 + 2.05x + 0.656y$ and $v = 0.754 - 2.18x - 2.05y$

"By definition, the acceleration of the fluid particle is the time derivative of the particle’s velocity, acceleration of a fluid particle a particle = $\dfrac{dV_{particle}}{dt}$"

We must use the chain rule and the equation becomes:
$particle(x, y, z, t) = \dfrac{dV_{particle}}{dt} = \dfrac{\partial V}{\partial t} + u(\dfrac{\partial V}{\partial t}) + v(\dfrac{\partial V}{\partial t}) + w(\dfrac{\partial V}{\partial t})$

The questions asks us for the expression for acceleration components $A_{x}$ and $A_{y}$, which are $u(\dfrac{\partial V}{\partial t})$ and $ v(\dfrac{\partial V}{\partial t})$

$A_{x}$ = $\dfrac{\partial u}{\partial t} + u(\dfrac{\partial u}{\partial x}) + v(\dfrac{\partial u}{\partial y}) + w(\dfrac{\partial u}{\partial z})$

$A_{y} = \dfrac{\partial v}{\partial t} + u(\dfrac{\partial v}{\partial x}) + v(\dfrac{\partial v}{\partial y}) + w(\dfrac{\partial v}{\partial z})$

Because t and z are not a variables in the velocity field the terms with those variables go to zero :

$A_{x} = u(\dfrac{\partial u}{\partial x}) + v(\dfrac{\partial u}{\partial y})$

$A_{y} = u(\dfrac{\partial v}{\partial x}) + v(\dfrac{\partial v}{\partial y})$

## Analysis:

```python
x , y = sym.symbols('x y') #find the partial derivatives
u_x = diff(x*2.05 + 0.656*y + 1.85, x)
u_y = diff(1.85 + 2.05*x + 0.656*y, y)

v_x = diff(0.754 - 2.18*x - 2.05*y, x)
v_y = diff(0.754 - 2.18*x - 2.05*y, y)
```

```python
%%render

x = -1
y = 3

A_x = (1.85 + 2.05*x + 0.656*y)*(u_x) + (0.754 - 2.18*x - 2.05*y)*(u_y)
A_y = (1.85 + 2.05*x + 0.656*y)*(v_x) + (0.754 - 2.18*x - 2.05*y)*(v_y)
```

## Discussion:
By applying the relation between acceleration and velocity as acceleration being the derivative of velocity, and the components of the acceleration being the partial derivative of the velocity field the acceleration in the $x$ component is $1.515$, and the acceleration in the $y$-component is $2.739$.
