1. Go to https://docs.anaconda.com/free/anaconda/install/windows/
if you are using a Mac computer or Linux, choose the right options from the left panel on the above page. 

2. Follow the steps on the webpage, and download the windows installer. In the right corner you see the downloaed installer.
