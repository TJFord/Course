t is the original time, and t + dt is a later time after t when the particle has deformed from its rectangular shape.

At a later time, the lower left corner coordinates become: $(x + u dt, y) = (x + (a + by) dt, y)$

Shear strain rate : half of the rate of decrease of the angle between two initially perpendicular lines that intercept at a point, originally the angle is 90 degress and it decreases.

Shear stress rate in Cartesian coordinates : $\epsilon_{xy} = (\dfrac{1}{2})(\dfrac{\partial u}{\partial y} + \dfrac{\partial v}{\partial x})$


## Analysis:
a) from the fundamental definition: $\epsilon_{xy} = (\dfrac{1}{2})$(d(change in angle between dx and dy))

We can apply trigonometric equations to find $\theta$, but we must find the x length of the triangle, we can do that by taking the coordinate of the upper left corner, and plug in the y-coordinate into the y variable of the x coordinate, and then subtract the lower left coordinate from that.

$(x + ( a + b ( y + dy) dt) - (x + ( a + by) dt) = (x + adt + bydt + bdydt) - (x + adt + bydt) = bdydt$

The hypotenuse would be dy, and so would be the y-coordinate length.

Because the deformation changes the x-side of the upper left coordinate we have a new angle $\theta$ and the difference of the angles ($\alpha$) can be found with trigonometic equations assuming that the change in x is opposite $\alpha$, and dy is both the hypotenuse and y.

$\alpha = {tan}^{-1}(bdydt / dt)$
> $= {tan}^{-1}(bdt)$
> $ bdt$

To calculate the new angle after deformation we can subtract $\alpha$ from the original angle of 90 degress to find $\theta$.

$\theta = \dfrac{\pi}{2} - bdt$

To find the shear stress rate we must subtract the new angle $\theta$ from the original angle

$\epsilon_{xy} = (\dfrac{-1}{2})(\dfrac{d[(\dfrac{\pi}{2} - bdt) - (\dfrac{\pi}{2})]}{dt})$
> $= (\dfrac{-1}{2})(\dfrac{d(dbdt)}{dt})$

> $= (\dfrac{1}{2})(b)$

> $= \dfrac{b}{2}$

b) The equation for the shear strain rate in cartesian coordinates : 

$\epsilon_{xy} = (\dfrac{1}{2})(\dfrac{\partial u}{\partial y} + \dfrac{\partial v}{\partial x})$

> $= (\dfrac{1}{2})(\dfrac{\partial a + by}{\partial y} + \dfrac{\partial (0)}{\partial x})$

> $= (\dfrac{1}{2})(b + 0)$

> $= \dfrac{b}{2}$

## Discussion: 
The shear strain rate from the fundamental definition was $\dfrac{b}{2}$, which is the same as the shear strain rate from the equation for cartesian coordinates.
