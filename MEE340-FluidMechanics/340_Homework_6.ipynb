

## Solution:
Given the information about the wind, and wind turbine, the mechanical energy of air per unit mass, the power generation potential of a wind turbine, and the actual electric power generation are to be determined.

## Assumptions:
The power potential of the wind is its kinetic energy, $e_{mech} = ke = \dfrac{V_{wind}^2}{2}$, per unit mass and $\dot{m}e_{mech}$ for a given mass flow rate.

## Analysis:

```python
%%render param

V_wind = 10*(u.m/u.s) #velocity of wind
D_blades = 70*u.m 
n = 30/100 #percent
rho_air = 1.25*(u.kg/u.m**3)
pi = math.pi
```

```python
%%render long
e_mech = (V_wind**2)/2
```

```python
print(e_mech.to('kJ/kg'))
```

$\dot{m} = \rho (V)(A)$

The power generation potential: $\dot{W}_{max} = \dot{E}_{mech} = e_{mech}(\dot{m})$

```python
%%render long
A = (pi*(D_blades**2))/4
m_dot = rho_air*V_wind*A
W_max = m_dot*e_mech
```

```python
print(W_max.to('kW'))
```

The actual electric power generation is given by the multiplication of the efficiency and power generation potential: $\dot{W}_{elect} = (\eta_{wind-turbine})(\dot{W}_{max})$

```python
%%render
W_elect = n*W_max
```

```python
print(W_elect.to('kW'))
```

## Discussion:
By using the equation from the book for kinetic energy, maximum power generated, and electric power, the mechanical energy of air per unit mass is $e_{mech} = 0.05 \dfrac{kJ}{kg}$, the power generation potential of the wind turbine is $\dot{W}_{max} = 2405.3 kW$, and the actual electric power generation is $\dot{W}_{elect} = 721.6 kW$.
