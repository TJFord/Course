Find $F_x=?$

## thought process

the problem asks to find the horizontal force $F_x$ of the fluid on the walls, then momentum equation should be used as it relates force to change of the momumtum. 

1. Draw a control volume as shown above, label with all the forces in x direction (weight is not shown as it is not in x direction)
2. Apply momentum equation to the control volume
$$\sum \vec{F}=\frac{d}{dt}\int_{cv} \rho \vec{V}d\vee+\int_{cs}\rho \vec{V}(\vec{V}\cdot\vec{n})dS$$
as we are only interested in the x direction, so we only consider the x component:
$$\sum F_{x,i}=\frac{d}{dt}\int_{cv} \rho u d\vee+\int_{cs}\rho u(\vec{V}\cdot\vec{n})dS$$
where $u$ is the x component velocity $\vec{V}$ (see how the equation changes)

3. Left hand side
$$\sum F_{x,i} = P_1A_1+P_2A_2 + F_x$$

4. right hand side
\begin{align}  
RHS &=\underbrace{\frac{d}{dt}\int_{cv} \rho ud\vee}_{=0 \text{ steady state}}+\int_{s_1}\rho u(\vec{V}\cdot\vec{n})dS+\int_{s_2}\rho u(\vec{V}\cdot\vec{n})dS+\underbrace{\int_{{other surfaces}}\rho u(\vec{V}\cdot\vec{n})dS}_{=0, \vec{V}\text{is zero on the side wall}}\\
&=\int_{s_1}\rho u(\vec{V}\cdot\vec{n})dS+\int_{s_2}\rho u(\vec{V}\cdot\vec{n})dS\\
&=\beta_1\int_{s_1}\rho V_1(V_1\vec{i}\cdot(-\vec{i})dS+\beta_2\int_{s_2}\rho (-V_2)(-V_2\vec{i}\cdot(-\vec{i})dS \quad\quad\text{($\beta$ is used as the velocity is non-uniform)}\\
&=-\beta_1\rho V_1^2A_1-\beta_2\rho V_2^2A_2
\end{align}


5. combine step 3 and 4, we get
$$P_1A_1+P_2A_2 + F_x=-\beta_1\rho V_1^2A_1-\beta_2\rho V_2^2A_2$$
$$ F_x=-\beta_1\rho V_1^2A_1-\beta_2\rho V_2^2A_2-P_1A_1-P_2A_2 $$
this is the force applied on the fluid by the duct. The force of fluid on the walls will be 
$F_{x, on duct}=-F_x = \beta_1\rho V_1^2A_1+\beta_2\rho V_2^2A_2+P_1A_1+P_2A_2 $

## what if the gaged pressure is non-uniform? 
$$F_{x, on duct}=-F_x = \beta_1\rho V_1^2A_1+\beta_2\rho V_2^2A_2+\int_{s_1}P_1dA+\int_{s_2}P_2dA $$

The only changes is the integral for the pressure.
