we need to calcuate mass and momentum flux on each surface. By definition, the flux is defined as some properties flowing in/out through a surface, $$\phi =\int \rho b \vec{u}\cdot \vec{n}dA$$. Take $b=1$ we get mass flux, and take $b=\vec{u}$ we get momentum flux. 
$$\begin{align}
\dot{m} = \int \rho (\vec{u}\cdot \vec{n})dA=\rho\int  (\vec{u}\cdot \vec{n})dA \\
\dot{\mathbf{P}} = \int \rho \vec{u}(\vec{u}\cdot \vec{n})dA = \rho\int  \vec{u}(\vec{u}\cdot \vec{n})dA
\end{align}$$

```python
import sympy as sp # use symbolic python taking derivative
import handcalcs.render # render
import numpy as np # numerical python library
import matplotlib.pyplot as plt
u, A,x,y,rho,h= sp.symbols('u A x y rho h') # define it as symbols
sp.init_printing()
#sp.init_session()
```

```python
%%render
u = sp.Matrix([A*x, -A*y, 0]) #velocity vector
n1 = sp.Matrix([0,-1,0]) # surface norm -j
n2 = sp.Matrix([1,0,0]) # surface norm i
uDotn1 = u.dot(n1) # this is dot(u, n1)
uDotn2 = u.dot(n2) # this is dot(u, n2)
mFlux1 = sp.integrate(rho*uDotn1,(x,0,h)) # [note:mass flux in surface 1]
mFlux1
mFlux2 = sp.integrate(rho*uDotn2,(y,0,h)) # mass flux in surface 2
# 1st surface
uuDotn1_x = u[0]*uDotn1 # u[0] is u_x
uuDotn1_y = u[1]*uDotn1 # u[1] is u_y
PFlux1_x = sp.integrate(rho*uuDotn1_x,(x,0,h)) # momentum in x direct on surface 1
PFlux1_y = sp.integrate(rho*uuDotn1_y,(x,0,h)) # momentum in y direct on surface 1
# second surface
uuDotn2_x = u[0]*uDotn2
uuDotn2_y = u[1]*uDotn2
PFlux2_x = sp.integrate(rho*uuDotn2_x,(y,0,h)) # momentum in x direct on surface 2
PFlux2_y = sp.integrate(rho*uuDotn2_y,(y,0,h)) # momentum in y direct on surface 2
```
