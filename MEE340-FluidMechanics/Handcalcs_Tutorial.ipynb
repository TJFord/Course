### The basic and default use of render is as follows:

- Write “%%render” 
- Create the variables with their values
- Create the formula

```python
%%render

x = 5
y = 6

z= x+y
```

### To make the output display the equation before the answer, leave a blank space after the name of the function and the equal sign when writing the equation

```python
%%render

x = 5
y = 6

z = x+y
```

### This default output is the same as when you add the word short after “%%render”, this makes the output appear in a single line

```python
%%render short

x = 5
y = 6

z = x+y
```

### When you substitute short with long, the output will appeared spread out into multiple lines


```python
%%render long

x = 5
y = 6

z = x+y
```

### Also, you can use symbolic after “%%render” to have the output of the equation only be the variables and not the mathematical result

```python
%%render symbolic

x = 5
y = 6

z = x+y
```

### In contrast by using param after “%%render”, the output will only display the mathematical result and not the variables

```python
%%render param

x = 5
y = 6

z = x+y
```

### To create a variable with a subscript write the name of the variable followed by an underscore and then the subscript: F_x


```python
%%render

m = 5
a = 6

F_x = m*a
```

### To specify the amount of decimals places write the number of decimal places after “%%render”

```python
%%render 3

x = 5.4
y = 6.3

z = x+y
```

### You can display certain symbols by typing key words which can be found on a table in : https://github.com/connorferster/handcalcs

```python
%%render 3

mu = 5.4
theta = 6.3

alpha = mu+theta
```

### You can also use complex number in your calculations:

```python
%%render 3

x = 5.4j
y = 6.3

z = x+y
```

### To add comments first write a pound, followed by a sentence:

This works in a text cell, which must be in Markdown and not Code.

```python
%%render 3

x = 5.4j
y = 6.3

z = x+y #this is the comment for a Code cell
```
