

## Solution: 
Given a velocity field, the pressure field as a function of $x$ and $y$ must be calculated.

## Assumptions:
The flow is steady, incompressible and two-dimensional in the $x-y$ plane. Also, $a$ is a constant in the velocity field, $V = (u, v) = (-ax^2)\vec{i} + (2axy)\vec{j}$

## Analysis: 
Begin by checking the continuity equation:
\begin{equation*}
\dfrac{\partial u}{\partial x} + \dfrac{\partial v}{\partial y} + \dfrac{\partial w}{\partial z} = 0
\end{equation*}

The first term corresponds to $-2ax$ as it is in the x-direction, the second term corresponds to $2axy$ because it is in the y-direction, and the third term goes to zero because there is no z-direction in this two-dimensional problem.

This proves that continuity is satisfied.

Next, check the x-component of the Navier-Stokes equation:
\begin{equation*}
\rho (\dfrac{\partial u}{\partial t} + u\dfrac{\partial u}{\partial x} + v\dfrac{\partial u}{\partial y} + w\dfrac{\partial u}{\partial z}) = -\dfrac{\partial P}{\partial x} + \rho g_x + \mu( \dfrac{\partial^2 u}{\partial x^2} + \dfrac{\partial^2 u}{\partial y^2} + \dfrac{\partial^2 u}{\partial z^2})
\end{equation*}

Because the flow is steady the first partial derivative goes to zero, the term $\rho g_x$ also goes to zero because gravity does not act in the x-direction, and the equation becomes

\begin{equation*}
\dfrac{\partial P}{\partial x} = -2\rho a^2 x^3 -2\mu a
\end{equation*}

which is the x momentum, and is satisfied if a pressure field is generated.

for the y momentum the equation is:

\begin{equation*}
\rho (\dfrac{\partial v}{\partial t} + u\dfrac{\partial v}{\partial x} + v\dfrac{\partial v}{\partial y} + w\dfrac{\partial v}{\partial z}) = -\dfrac{\partial P}{\partial y} + \rho g_y + \mu( \dfrac{\partial^2 v}{\partial x^2} + \dfrac{\partial^2 v}{\partial y^2} + \dfrac{\partial^2 v}{\partial z^2})
\end{equation*}

This equation becomes:

\begin{equation*}
\dfrac{\partial P}{\partial y} = -2\rho a^2 x^2 y
\end{equation*}

Next, differentiate the x and y momentum equations:

\begin{equation*}
\dfrac{\partial^2 P}{\partial y \partial x} = \dfrac{\partial}{\partial y}(\dfrac{\partial P}{\partial x}) = 0 \qquad \dfrac{\partial^2 P}{\partial x \partial y} = \dfrac{\partial}{\partial x}(\dfrac{\partial P}{\partial y}) = -2\rho a^2 x^2
\end{equation*}

## Discussion:
Because the terms don't match $P$, this is not a smooth function of $x$ and $y$, and a steady, incompressible, two-dimensional pressure field cannot be calculated.
