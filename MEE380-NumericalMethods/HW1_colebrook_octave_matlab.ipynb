Estimate the friction coefficient f in the following equation. Assuming the wall roughness (𝜖 = 0), Re = 
1e6. 

# Plan
To find the solution to the above equation we can convert it to find the insertction point between two curves $$g_1(f)=\dfrac{1}{\sqrt{f}}$$ and $$g_2(f)=-2.0\log(\dfrac{\epsilon/D}{3.7}+\dfrac{2.51}{Re\sqrt{f}})$$
so we can plot these two curves for a number of values for $f$, e.g, $f\in (0,0.05)$

```octave
%% 1. the anlytical solution is not availble, so we choose numerial approach based on graphics
% 2. define function g1(f)=1/sqrt(f), g2(f)=-2.0*log10(2.51/Re./sqrt(f));
% 3. plot both g1 and g2 on the same figure
% 4. the intersection point would be the solution to the orignal equation
% Jifu Tan, jifutan@niu.edu, 1/24/2024

clear;clc;               % clear all variables, clear the command window
f = 0.01:0.0001:0.015;   % define f starting with 0.01 and stopping at 0.015 with an increament of 0.0001
g1 = 1./sqrt(f);         % define function g_1
Re = 1e6;                % Re
g2 = -2.0*log10(2.51/Re./sqrt(f));  % define function g_2
figure;                 % start a new figure
plot(f,g1,'b','linewidth',3);           % plot g1 using color blue and linewidth 3
hold on                  % hold on the axis so that the second plot can be shown as well
plot(f,g2,'r')            % plot the 2nd curve for g2
legend('g1','g2')         % label each curve
grid on                   % put a grid on to the figure;
xlabel('f')              % label x axis
ylabel('g')              % label y axis
```

# Results

Through visual check, we can see that these two curves meet between 0.011 and 0.012. Let us pick the middle value as the approximation $f=0.0115$, which gives us the friction factor.

Obviously, the approximation using visual check cannot give us precise results if a high precision is required. To find the better approximations, we will use other root finding methods that will be discussed in future lectures.
