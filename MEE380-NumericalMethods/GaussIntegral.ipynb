## Gauss quadrature/integral
<div>

</div>
Trapezoidal rule $I\approx \int_a^b f(x)dx=\dfrac{f(a)+f(b)}{2}(b-a)=\underbrace{\dfrac{(b-a)}{2}}_{w_1}\underbrace{f(a)}_{f(x_1)}+\underbrace{\dfrac{(b-a)}{2}}_{w_2}\underbrace{f(b)}_{f(x_2)}$

For gauss integral, we consider $$I=\int_{-1}^{1}f(x)dx=\sum_{i=1}^nw_if(x_i)$$
$n$ is the number of gauss points for integration

## question

Consider $n=2$, how can we find the values for weight $w_1, w_2$ associated with two gauss points $x_1, x_2$?

## strategy
if gauss integral formula works, it should work for simple polynomial functions. For $n=2$, we can try function $f(x)={1, x, x^2, x^3}$, such that
    \begin{align}
    \int_{-1}^11dx&=w_1f(x_1)+w_2f(x_2)=w_1+w_2=2\\
    \int_{-1}^1xdx&=w_1f(x_1)+w_2f(x_2)=w_1x_1+w_2x_2=0\\
    \int_{-1}^1x^2dx&=w_1f(x_1)+w_2f(x_2)=w_1x_1^2+w_2x_2^2=\dfrac{2}{3}\\
    \int_{-1}^1x^3dx&=w_1f(x_1)+w_2f(x_2)=w_1x_1^3+w_2x_2^3=0
    \end{align}
only 4 equations are needed for 4 unknowns. 

In summary, we can find $w_1=w_2=1, x_1=-\sqrt{\dfrac{1}{3}}, x_2=\sqrt{\dfrac{1}{3}}$

## discusssions
1. Gausss integral works for the domain [-1, 1], what should we do if the integral bound is [a,b]? we need to reformulate the integral through change of varialbes. $u(x)=kx+c$ such that $u(a)=-1, u(b)=1$

## example
integrate $$I=\int_1^32x^3-x^2dx$$ using two gauss points

### wrong solution
\begin{align}
I=\int_1^32x^3-x^2dx&=w_1f(x_1)+w_2f(x_2)\\
&=f(-\sqrt{1/3})+f(\sqrt{1/3})
\end{align}
because the integral domain is not within $[-1,1]$

### correct solution
1. find the mapping for $u(x)=kx+c$ such that $u(1)=-1, u(3)=1$
$u(x)=x-2$ 
then we can write $x=u+2$, $dx=du$, the original integral becomes

\begin{align}
I=\int_1^32x^3-x^2dx&=\int_{-1}^1\underbrace{2(u+2)^3-(u+2)^2}_{f(u)}du=w_1f(u_1)+w_2f(u_2)\\
&=f(-\sqrt{1/3})+f(\sqrt{1/3})
\end{align}
