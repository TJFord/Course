# Solving Linear Systems using gauss elimination with partial pivoting

Solving
\begin{align}
2x_1+5x_2+x_3&=1 \\
6x_1+2x_2-3x_3&=9\\
x_1+2x_2+4x_3&=-3
\end{align}

## Strategy

We use row reduction to solve the linear system. It comprises two steps: **Forward Elimination** and **Backward Substitution**. **Forward Elimination** is to eliminate the lower triangular elements so that they become zeros. e.g., something like this

$Aug = \begin{bmatrix}
a_{11} & a_{12} & a_{13} & b_1\\
0 & a_{22} & a_{23} & b_2\\
0 & 0 & a_{33} & b_3
\end{bmatrix}$

The detailed steps can be found in the following example.

```octave
clear;clc;
x = zeros(3,1)

for i=1:10

x(1)=(9+3*x(3)-2*x(2))/6
x(2)=(1-x(3)-2*x(1))/5
x(3)=(-3-2*x(2)-x(1))/4
end

```
